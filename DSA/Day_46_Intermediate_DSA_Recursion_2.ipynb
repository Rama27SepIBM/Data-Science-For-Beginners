Q1. Implement Power Function

Implement pow(A, B) % C.
In other words, given A, B and C, Find (AB % C).

Note: The remainders on division cannot be negative. In other words, make sure the answer you return is non-negative.

```python
class Solution:
    # @param A : integer
    # @param B : integer
    # @param C : integer
    # @return an integer
    def pow(self, A, B, C):
        return self.Expo(A,B,C)

    def Expo(self, A,B,C):
        if B>1:
            val =(self.Expo(A,B//2,C))
            if B%2==0:
                return (val*val)%C
            else:
                return ((val*val)%C * A%C)%C
        if B==1:
            return A%C
        else:
            return 1%C
```

Q1. Is magic?

Given a number A, check if it is a magic number or not.

A number is said to be a magic number if the sum of its digits is calculated till a single digit recursively by adding the sum of the digits after every addition. If the single digit comes out to be 1, then the number is a magic number.

```python
def calSumOfdigit(A):
    if A == 0 :
        return A
    return calSumOfdigit(A//10) + A % 10
class Solution:
    # @param A : integer
    # @return an integer
    def solve(self, A):
        while A >= 10:
            A = calSumOfdigit(A)
        # Take modulo to 9 as it the greatest number of the single digit
        if A % 10 == 1:
            return 1
        else:
            return 0
```

Q2. Kth Symbol - Easy

On the first row, we write a 0. Now in every subsequent row, we look at the previous row and replace each occurrence of 0 with 01, and each occurrence of 1 with 10.

Given row number A and index B, return the Bth indexed symbol in row A. (The values of B are 0-indexed.).

```python
class Solution:

    # @param A : integer

    # @param B : integer

    # @return an integer

    def solve(self, A, B):

        row = "0"

        for i in range(1, A):

            new_row = ""

            for symbol in row:

                if symbol == "0":

                    new_row += "01"

                else:

                    new_row += "10"

            row = new_row

        return row[B]
```

Q3. Kth Symbol - Hard

On the first row, we write a 0. Now in every subsequent row, we look at the previous row and replace each occurrence of 0 with 01, and each occurrence of 1 with 10.

Given row number A and index B, return the Bth indexed symbol in row A. (The values of B are 0-indexed.).

```python
class Solution:
    # @param A : integer
    # @param B : long
    # @return an integer
    def solve(self, A, B):
        if B==0:
            return 0
        x=self.solve(A-1,B//2)
        if B%2==0:
            return x
        return 1-x
```
