Q2. Length of longest consecutive ones

Given a binary string A. It is allowed to do at most one swap between any 0 and 1. Find and return the length of the longest consecutive 1â€™s that can be achieved.


Input Format

The only argument given is string A.

```python
class Solution:
    # @param A : string
    # @return an integer
    def solve(self, A):
        cnt_one = 0  # To count all 1's in the string
        n = len(A)
        max_cnt = 0
        for x in A:
            if x=='1':
                cnt_one+=1.   # To store cumulative 1's
        left = [0]*n
        right = [0]*n
        if A[0]=='1':
            left[0] = 1
        if A[n-1]=='1':
            right[n-1] = 1
        for i in range(1, n):
            if A[i]=='1':
                left[i] = left[i-1] + 1
        for i in range(n-2, -1, -1):
            if A[i]=='1':
                right[i] = right[i + 1] + 1
                cnt = 0
                max_cnt = 0;
        for i in range(n):
            max_cnt = max(max_cnt, max(right[i], left[i]))
        for i in range(1, n-1):
            if A[i]=='0':
                su = left[i-1] + right[i+1]
                cnt = su
                if su < cnt_one:
                    cnt += 1
                max_cnt = max(max_cnt, cnt)
                cnt = 0
        return max_cnt
```

Q3. Count Increasing Triplets

You are given an array A of N elements. Find the number of triplets i,j and k such that i<j<k and A[i]<A[j]<A[k]

```python
class Solution:
    # @param A : list of integers
    # @return an integer
    def solve(self, A):
        n = len(A)
        count = 0
        
        for i in range(n):
            l = 0
            r = 0
            for j in range(i):
                if A[j] < A[i]:
                    l+=1
            for j in range(i+1,n):
                if(A[i] < A[j]):
                    r+=1
            count += (l * r)
        return count
        # Time Complexity : O(N^2)  Space Complexity : O(1)
```

Q4. Josephus Problem - 2

There are A people standing in a circle. Person 1 kills their immediate clockwise neighbour and pass the knife to the next person standing in circle. This process continues till there is only 1 person remaining. Find the last person standing in the circle.

```python
class Solution:
    # @param A : integer
    # @return an integer
    def solve(self, A):
        n=1
        while A>=n:
            n=n*2
        n=n//2
        return 1+2*(A-n)
```

Q4. Multiple left rotations of the array

Given an array of integers A and multiple values in B, which represents the number of times array A needs to be left rotated.

Find the rotated array for each value and return the result in the from of a matrix where ith row represents the rotated array for the ith value in B.

```python
class Solution:
    # @param A : list of integers
    # @param B : list of integers
    # @return a list of list of integers
    def reverse(self, arr , start, end):
        while start<end:
            arr[start] , arr[end] = arr[end], arr[start]
            start += 1
            end -= 1
        return arr
    def reverse_ktimes(self,arr,k):
        n = len(arr)
        k = k%n
        self.reverse(arr, 0, k-1)
        self.reverse(arr, k,n-1)
        self.reverse(arr, 0, n-1)
        return arr
    def solve(self, A, B):
        output = []
        for i in B:
            array = A.copy()
            output.append(self.reverse_ktimes(array,i))
        return output
```
