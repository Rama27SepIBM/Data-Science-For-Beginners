Q1. Divisible by 5 and 7

Write a program that returns the list of elements that are present in the given list and are divisible by 5 and 7.

Input Format:

The only argument is the given list of elements
Output Format:

Return a list consisiting of the elements that are divisible by 5 and 7 from the given list.
Sample Input:

[23, 34, 25, 35, 66, 70, 14]
Sample Output:

[35, 70]
Sample explanation:

Only 35 and 70 are divisible by 5 and 7 from the list [23,34,25,35,66,70,14] 
therefore the list consisting of only these two elements is returned.

```python
def divisible(lst):
    ''' input:lst-List of elements in integer format
         output:Return the list of elements divisible by 5 and 7'''
    
    elements=[]
    # YOUR CODE GOES HERE
    for i in lst:
        if i %5==0 and i %7==0:
            elements.append(i)

    
    return elements
```

Q2. Second Largest

You are given an integer array A. You have to find the second largest element/value in the array or report that no such element exists.


Problem Constraints
1 <= |A| <= 105

```python
class Solution:
    # @param A : list of integers
    # @return an integer
    def solve(self, A):
        if len(A) < 2:
            return -1
        else:
            B = set(A)
            b = max(A)
            B.remove(b)
            return max(B)
```

Q1. First and Last Index

Given a sorted integer array A, and an integer B. Find the first and last index of B in A.
It is guaranteed that B exists in A. Return an array C of size 2, where C[0] is the first index of B in A and C[1] is the last index of B in A.

Note - Indexing of A starts from 0. If the element only exists once, the values of both elements in the C list should be the same.

Problem Constraints
1 <= |A| <= 105

```python
class Solution:
    # @param A : list of integers
    # @param B : integer
    # @return a list of integers
    def solve(self, A, B):
        c=[-1,-1]
        count=0
        for i in range(len(A)):
            if A[i]==B:
                if count<1:
                    c[0]=i
                    c[1]=i
                    count+=1
                elif count>=1:
                    c[1]=i
        return c
```

Q2. Ceil and Floor

You are given a sorted array of A and an integer B. Return an array of size 2, where the first element is the ceil of B and the second element is the floor of B.

Ceil is the largest number greater than or equal to B and is present in the array A.
Floor is the largest number smaller than or equal to B and is present in the array A.


Problem Constraints
1 <= |A| <= 105

```python
class Solution:
    # @param A : list of integers
    # @param B : integer
    # @return a list of integers
    def solve(self, A, B):
        C = [-1,-1]
        if B in A:
            C[0] = B
            C[-1] = B
        else:
            for i in A:
                if B >= i:
                    C[-1] = i
                if B <= i:
                    C[0] = i
                    break
        return C
```

Q3. Inverse of the Array

You are given an integer array A. Now your task is to find the inverse of A.
Now, the inverse of the array is A will be an array in which we change the
positions of the values as their indices and indices as values.

So, array A = [2, 0, 1]
- Now 2 is at index 0. So, place 0 at index 2.
- 0 is at index 1. So, place 1 at index 0.
- 1 is at index 2. So, place 2 at index 1.

So, the inverse of A will be [1, 2, 0]


Problem Constraints
1 <= |A| <= 105
0 <= A[i] < |A| (All elements are distinct)

```python
class Solution:
    # @param A : list of integers
    # @return a list of integers
    def solve(self, A):
        B = [0]*len(A)
        for i in range(len(A)):
            B[A[i]] = i

        return B
```

Q6. Check if Array is Sorted

Given an array A, check if it is sorted in ascending order or not.

Problem Constraints
1 <= Ai <= 109
1 <= |A| <= 105


Input Format
The first and only argument contains an integer array A.


```python
class Solution:
    # @param A : list of integers
    # @return an integer
    def solve(self, A):
        if A == sorted(A):
            return 1
        else:
            return 0
```
