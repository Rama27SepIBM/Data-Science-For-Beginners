Q1. Greatest Common Divisor

Given 2 non-negative integers A and B, find gcd(A, B)

GCD of 2 integers A and B is defined as the greatest integer 'g' such that 'g' is a divisor of both A and B. Both A and B fit in a 32 bit signed integer.

Note: DO NOT USE LIBRARY FUNCTIONS.



Problem Constraints
0 <= A, B <= 109

```python
def Gcd(x,y):
    if y==0:
        return x
    else:
        return Gcd(y,x%y)
class Solution:
    # @param A : integer
    # @param B : integer
    # @return an integer
    def gcd(self, A, B):
        
            return Gcd(A,B)
            
```

Q3. Number Of Open Doors

Given an integer A, which denotes the number of doors in a row numbered 1 to A. All the doors are closed initially.

A person moves to and fro, changing the states of the doors as follows: the person opens a door that is already closed and closes a door that is already opened.

In the first go, he/she alters the states of doors numbered 1, 2, 3, … , A.
In the second go, he/she alters the states of doors numbered 2, 4, 6 ….
In the third go, he/she alters the states of doors numbered 3, 6, 9 …
This continues till the A'th go in, which you alter the state of the door numbered A.

Find and return the number of open doors at the end of the procedure.



Problem Constraints
1 <= A <= 109

```python
class Solution:
    # @param A : integer
    # @return an integer
    def solve(self, A):
        return int(math.sqrt(A))
```

Q4. Find All Primes

Given an integer A. Find the list of all prime numbers in the range [1, A].


Problem Constraints
1 <= A <= 106

```python
class Solution:
    # @param A : integer
    # @return a list of integers
    def solve(self, A):
        primes = [True] * (A-1)
        for i in range(2, int(A**0.5)+1):
            if primes[i-2]:
                for j in range(i*i, A+1, i):
                    primes[j-2] = False
        result = []
        for i in range(2, A+1):
            if primes[i-2]:
                result.append(i)
        return result
```

Q1. Delete one

Given an integer array A of size N. You have to delete one element such that the GCD(Greatest common divisor) of the remaining array is maximum.

Find the maximum value of GCD.



Problem Constraints

2 <= N <= 105

```python
from math import gcd

class Solution:
    # @param A : list of integers
    # @return an integer
    def solve(self, A):
        n = len(A)
        prefix_gcds = [0] * n
        suffix_gcds = [0] * n
        prefix_gcds[0] = A[0]
        suffix_gcds[-1] = A[-1]


        for i in range(1, n):
            prefix_gcds[i] = gcd(prefix_gcds[i-1], A[i])
        for i in range(n-2, -1, -1):
            suffix_gcds[i] = gcd(suffix_gcds[i+1], A[i])
        max_gcd = -1
        for i in range(n):
             if i == 0:
                 max_gcd = max(max_gcd, suffix_gcds[1])
             elif i == n-1:
                 max_gcd = max(max_gcd, prefix_gcds[n-2])
             else:
                  max_gcd = max(max_gcd, gcd(prefix_gcds[i-1], suffix_gcds[i+1]))
        return max_gcd
```

Q2. IsPrime

Given an Integer A. Return 1 if A is prime and return 0 if not.


Problem Constraints
1 <= A <= 1012

```python
class Solution:
    # @param A : long
    # @return an integer
    def solve(self, A):
        if A <= 1:
            return 0
        for i in range(2, int(A**0.5)+1):
            if A % i == 0:
                return 0
        return 1
```
