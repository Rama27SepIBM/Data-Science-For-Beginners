```python
class Solution:
    # @param A : list of integers
    # @return a list of integers
    def solve(self, A):
        ans=0
        B=0
        for i in range(len(A)):
            curr_sum=0
            j = i
            while j <(len(A)):
                curr_sum = curr_sum + A[j]
                if curr_sum < B:
                    ans +=1
                    j +=1
        return ans
        
```
