Q 3 
Given string A count the number of uppercase letters in the string A

```python
def main():
    # YOUR CODE GOES HERE
    T = int(input())
    for i in range (T):
        N=input()
        count=0
        for i in N:
            if i.isupper():
                count+=1
        print(count)
                
    # Please take input and print output to standard input/output (stdin/stdout)
    # E.g. 'input()/raw_input()' for input & 'print' for output

    return 0

if __name__ == '__main__':
    main()
```
