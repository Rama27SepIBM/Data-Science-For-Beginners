```python
class Solution:
    # @param A : list of integers
    # @param B : integer
     # @return an long
    def solve(self, A, B):
        n = len(A)
        count = 0
        prefix_sum = [0] * (n + 1)
        for i in range(1, n + 1):
            prefix_sum[i] = prefix_sum[i - 1] + A[i - 1]

            mod_count = [0] * B
        for i in range(n + 1):
            count += mod_count[prefix_sum[i] % B]
            mod_count[prefix_sum[i] % B] += 1
        return count
```

```python
class Solution:
    # @param A : string
     # @return an long
    def solve(self, A):
        def sum_of_numbers_in_string(A):
            sum = 0
            num = 0
            sign = 1
            for char in A:
                if char.isdigit():
                    num = num * 10 + int(char)
                else:
                    #print(sign,num)
                    sum += sign * num
                    num = 0
                if char == '-':
                    sign = -1
                else:
                    sign = 1
            sum += sign * num
            return sum
        return sum_of_numbers_in_string(A)
```
