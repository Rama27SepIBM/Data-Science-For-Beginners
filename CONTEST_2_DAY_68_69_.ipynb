```python
class Solution:
    # @param A : list of integers
    # @return an integer
    def solve(self, A):
        n = len(A)
        chunks = []
        sorted_chunks = []
        chunk_size = 1
        while sum(len(c) for c in chunks) < n:
            chunk = A[chunk_size-1::chunk_size]
            chunks.append(chunk)
            sorted_chunk = sorted(chunk)
            sorted_chunks.append(sorted_chunk)
            chunk_size += 1
        sorted_A = [x for chunk in sorted_chunks for x in chunk]
        return sorted_A == A
```
